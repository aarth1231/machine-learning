```python
# Importing required libraries for data manipulation and mathematical operations
import pandas as pd
import math
```

```python
#Loading the heart disease dataset
df=pd.read_csv("heart_disease.csv")
df.head()
```

```python
#columns in the dataframe
df.columns
```

```python
#finding columns which have empty data
columns_with_missing = df.columns[df.isnull().any()].tolist()
print(columns_with_missing)
```

```python
#find columns with missing data for yes_no and created a dictionary to convert yes_no into 1_0
missing_yes_no= ['Smoking', 'Family Heart Disease', 'Diabetes','High Blood Pressure', 'Low HDL Cholesterol', 'High LDL Cholesterol']
yes_no_map = {'Yes': 1, 'No': 0}
```

```python
# Defining Yes/No columns that need cleaning and filling
# Stripping whitespace, capitalizing, filling missing values with mode
# Then mapping Yes->1 and No->0 for machine learning
for col in missing_yes_no:
     df[col]=df[col].str.strip().str.capitalize()
     mode_value = df[col].mode()[0]
     df[col] = df[col].fillna(mode_value)
     df[col] = df[col].map(yes_no_map) 
```

```python
#Filling missing Gender values with mode
# Then mapping Male->1 and Female->0 for machine learning
df['Gender'] = df['Gender'].fillna(df['Gender'].mode()[0])
df['Gender'] = df['Gender'].str.strip().str.capitalize()
df['Gender'] = df['Gender'].map({'Male': 1, 'Female': 0})
```

```python
# Filling missing values in all numerical columns using median
# Using math.floor to get whole number values
age_m= math.floor(df.Age.median())
df.Age=df.Age.fillna(age_m)
blood_pres=math.floor(df['Blood Pressure'].median())
df['Blood Pressure']=df['Blood Pressure'].fillna(blood_pres)
cholesterol_l= math.floor(df['Cholesterol Level'].median())
df['Cholesterol Level']=df['Cholesterol Level'].fillna(cholesterol_l)
bmi= math.floor(df.BMI.median())
df.BMI=df.BMI.fillna(bmi)
sleep_h=math.floor(df['Sleep Hours'].median())
df['Sleep Hours']=df["Sleep Hours"].fillna(sleep_h)
triglyceride_l=math.floor(df['Triglyceride Level'].median())
df['Triglyceride Level']=df['Triglyceride Level'].fillna(triglyceride_l)
fasting_b_s=math.floor(df['Fasting Blood Sugar'].median())
df['Fasting Blood Sugar']=df['Fasting Blood Sugar'].fillna(fasting_b_s)
crp_l=math.floor(df['CRP Level'].median())
df['CRP Level']=df['CRP Level'].fillna(crp_l)
homocyst_l=math.floor(df['Homocysteine Level'].median())
df['Homocysteine Level']=df['Homocysteine Level'].fillna(homocyst_l)
```

```python
#creating dummies with labelencoder
from sklearn.preprocessing import LabelEncoder
le_exercise_h=LabelEncoder()
le_alcohol_c=LabelEncoder()
le_stress_l=LabelEncoder()
le_sugar_c=LabelEncoder()
```

```python
df['exercise_n']=le_exercise_h.fit_transform(df['Exercise Habits'])
df['alcohol_n']= le_alcohol_c.fit_transform(df['Alcohol Consumption'])
df['stress_n']=  le_stress_l.fit_transform(df['Stress Level'])
df['sugar_n']=   le_sugar_c.fit_transform(df['Sugar Consumption'])
```

```python
# Dropping original categorical columns since we have their encoded versions
df=df.drop(['Exercise Habits','Alcohol Consumption','Stress Level','Sugar Consumption'],axis='columns')
df.head()
```

```python
# Mapping target column and Separating target variable y from feature dataframe df
df['Heart Disease Status'] = df['Heart Disease Status'].map({'Yes': 1, 'No': 0})
y=df['Heart Disease Status']
df = df.drop(['Heart Disease Status'], axis='columns')
```

```python
df.head()
```

```python
# Splitting data into 80% training and 20% testing
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(df,y,test_size=0.2)
```

```python
# Setting up Stratified KFold with 5 splits it ensures each fold has same Yes/No ratio as original data
from sklearn.model_selection import StratifiedKFold, cross_val_score
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
```

```python
# Training Support Vector Machine model on training data
# Running cross validation with Stratified KFold to check consistency
from sklearn.svm import SVC
model_svm=SVC()
model_svm.fit(x_train,y_train)
svm_scores   = cross_val_score(model_svm,x_train, y_train, cv=skf)
print("SVM CV Scores   :", svm_scores)
```

```python
# Training Random Forest model with 100 decision trees on training data
# Running cross validation with Stratified KFold to check consistency
from sklearn.ensemble import RandomForestClassifier
model_forest=RandomForestClassifier(n_estimators=100)
model_forest.fit(x_train,y_train)
forest_scores= cross_val_score(model_forest, x_train, y_train, cv=skf)
print("Forest CV Scores:", forest_scores)
```

```python
# Predicting Heart Disease on test set using both models
# Printing test accuracy for both models
svm_pred    = model_svm.predict(x_test)
forest_pred = model_forest.predict(x_test)
print("SVM    Test Score:", model_svm.score(x_test, y_test))
print("Forest Test Score:", model_forest.score(x_test, y_test))       
```
