```python
import pandas as pd
df=pd.read_excel('moblie_classifier.xlsx')
```

```python
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn import svm
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
```

```python
from sklearn.model_selection import train_test_split
x, y = df.drop('price_range', axis=1), df['price_range']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
models_para={    
    'logistic_regression' : {
        'model': LogisticRegression(solver='lbfgs',max_iter=1000),
        'params': {
            'C': [1,5,10]
        }
    },
     'random_forest': {
        'model': RandomForestClassifier(),
        'params' : {
            'n_estimators': [20,40,60,80,100]
        }
    },
    'svm': {
        'model': svm.SVC(gamma='auto'),
        'params' : {
            'C': [10,15,20,25],
            'kernel': ['rbf','linear']
        }  
    },
     'naive_bayes_gaussian': {
        'model': GaussianNB(),
        'params': {}
    },
     'decision_tree': {
        'model': DecisionTreeClassifier(),
        'params': {
            'criterion': ['gini','entropy'],

    }
}     
}
scores = []
```

```python
from sklearn.model_selection import RandomizedSearchCV
for model_name,mp in models_para.items():
     clf1 =  RandomizedSearchCV(mp['model'], mp['params'], cv=5, n_iter=2,return_train_score=False)
     clf1.fit(x_train, y_train)
     scores.append({
        'model': model_name,
        'search_type': 'RandomizedCV',
        'best_score': clf1.best_score_,
        'best_params': clf1.best_params_
    })
```

```python
from sklearn.model_selection import GridSearchCV
for model_name,mp in models_para.items():
    clf =  GridSearchCV(mp['model'], mp['params'], cv=3,n_jobs=-1, return_train_score=False)
    clf.fit(x_train, y_train)
    scores.append({
        'model': model_name,
        'best_score': clf.best_score_,
        'best_params': clf.best_params_
    })
```

```python
df = pd.DataFrame(scores,columns=['model','best_score','best_params'])
df
```
