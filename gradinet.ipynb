```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def gradient_descent(x,y):
    m=b=0
    iterations=1000
    n=len(x)
    learning_rate=0.008
    for i in range(iterations):
        ye= m*x+b
        cost=(1/n)* sum([val**2 for val in (y-ye)])
        md=-(2/n)*sum(x*(y-ye))
        bd=-(2/n)*sum(y-ye)
        m=m-learning_rate *md
        b=b-learning_rate*bd
        print ("m_curr{},b_curr{},cost{},iteration{}".format(m,b,cost,i))

    plt.scatter(x , y , color='red', marker='+',linewidths=5)
    plt.plot(x,ye,color='green')
    plt.show()
    return m, b
```

```python
x=np.array([1,2,3,4,5])
y=np.array([5, 7, 9, 11, 13])
gradient_descent(x, y)
```
